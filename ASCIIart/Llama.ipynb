```python
# Install dependencies
!pip install langchain-groq langchain_community gradio chromadb

# Import required libraries
import gradio as gr
from langchain_community.document_loaders import WebBaseLoader
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from langchain_groq import ChatGroq
import chromadb

# Initialize LLM (Replace with your API key)
llm = ChatGroq(model="llama-3.3-70b-versatile", api_key="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")


def generate_cold_email(job_url):
    """
    Takes a job posting URL, extracts job details, matches relevant projects from ChromaDB,
    and generates a cold email.
    """

    # Step 1: Scrape Job Description
    loader = WebBaseLoader(job_url)
    page_content = loader.load().pop().page_content

    # Step 2: Extract Job Details as JSON
    prompt_extract = PromptTemplate.from_template(
        """
        ### Scraped text from website:
        {page_content}

        ### Instruction:
        The scrapped text is from the careers page of a website.
        Your job is to extract the job posting and return them in JSON format containing keys:
        `role`, `experience`, `skills`, and `description`.
        Only return valid JSON.

        ### Valid JSON (No Preamble)
        """
    )
    chain_extract = prompt_extract | llm
    extracted_json = chain_extract.invoke({"page_content": page_content})

    json_parser = JsonOutputParser()
    job_details = json_parser.parse(extracted_json.content)


    # Step 4: Generate Cold Email
    prompt_email = PromptTemplate.from_template(
    """
    ### Job Description:
    {job_description}

    Write a professional cold email as **Vamshi**, a **Business Development Executive at Atliq**, an AI-driven software company.

    The email should:
    - Address the hiring manager professionally.
    - Highlight **Atliq’s expertise** in AI-powered automation and digital transformation.
    - Emphasize how Atliq can fulfill the job requirements efficiently.
    - Include **relevant project links** from {link_list} in a structured format.
    - End with a clear **call to action** for further discussion.

    **Output format:**
    - Use bullet points or highlights where necessary.
    - Make it engaging and easy to read.
    - Avoid unnecessary preamble.

    ### Cold Email (No Preamble):
    """
    )
    chain_email = prompt_email | llm
    cold_email = chain_email.invoke({
        "job_description": str(job),
        "link_list": links
    })

    return cold_email.content


# Build Gradio Interface
iface = gr.Interface(
    fn=generate_cold_email,
    inputs=gr.Textbox(label="Enter Job Post URL"),
    outputs=gr.Textbox(label="Generated Cold Email"),
    title="Cold Email Generator for Job Postings",
    description="Enter a job posting URL, and get an AI-generated cold email with relevant project links."
)

# Launch Web App
iface.launch(share=True)
```
